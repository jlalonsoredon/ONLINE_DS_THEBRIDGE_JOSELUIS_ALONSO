## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
array_strings = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]
duplicado = []
for i in range(len(array_strings)):
  palabra =array_strings[i].lower()
  for j in range(i+1, len(array_strings)):
    if palabra == array_strings[j]:
      if palabra not in duplicado:
        duplicado.append(palabra)


for palabra_duplicada in duplicado:
    indices = []
    for pos in range(len(array_strings)):
        if array_strings[pos].lower() == palabra_duplicada:
            indices.append(pos)

    print("Palabra:", palabra_duplicada, "- está en índices: ", indices)
`````

## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
array_strings = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

def encontrar_indices(valor_a_buscar):
    indices = []
    busqueda = valor_a_buscar.lower()

    for pos in range(len(array_strings)):
        if array_strings[pos].lower() == busqueda:
            indices.append(pos)

    return indices

print("Resultados:")
print(f"Indices de 'un': {encontrar_indices('un')}")
print(f"Indices de 'es': {encontrar_indices('es')}")
print(f"Indices de 'binario': {encontrar_indices('binario')}")
`````

## Ejercicio 11
Convierte el programa del ejercicio 9 en una función al que se le pasen como argumentos la librería, posicional, y el título, argumento keyword con valor "ninguno", por defecto.

`````python
def buscar_libro(libreria, titulo="ninguno"):
    encontrado = False

    for libro in libreria:
        if libro["titulo"].lower() == titulo.lower():
            print("Libro encontrado:")
            print(libro)
            encontrado = True
            break

    if not encontrado:
        print(f"El libro '{titulo}' no lo tengo, ¿mola?")

buscar_libro(libreria)
buscar_libro(libreria, titulo="1984")
`````

Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.

>```Python
>DNI= 55555555
